```python
def nev_atiro(nev):
    if len(nev.split())>1:
        if nev.lower() in simanevek:
            uj_nev=nev
        else:
            uj_nev=nev.split()[0][0]+"."
            for i in range(1,len(nev.split())):
                uj_nev=uj_nev+" "+nev.split()[i]
    else:
        uj_nev=nev
    return uj_nev.lower()
def betu_atiro(nev):
    uj_nev=nev.lower().replace("ý", "y").replace("ł","l").replace("ź","z").replace("ń","n").replace("ś","s").replace("ć","c").replace("ę","e").replace('ř','r').replace("ž","z").replace("ż","z").replace("š","s").replace("ç","c").replace("č","c").replace("ą","a").replace("ó","o")
    return uj_nev

```
